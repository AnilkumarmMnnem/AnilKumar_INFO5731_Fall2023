```python
# Write your code here

import requests
from bs4 import BeautifulSoup as bs
import re
```

```python
iphone_reviews=[]
for i in range(1,1000):
  ip=[]
  url="https://www.amazon.com/Apple-iPhone-11-64GB-Unlocked/product-reviews/B07ZPKF8RG/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews"+str(i)
  response = requests.get(url)
  soup = bs(response.content,"html.parser")
  reviews = soup.find_all("div",attrs={"class","a-row a-spacing-small review-data"})
  for i in range(len(reviews)):
    ip.append((reviews[i].text).replace('\n',''))
    iphone_reviews=iphone_reviews+ip
```

# **Question 2**

(30 points). Write a python program to **clean the text data** you collected above and save the data in a new column in the csv file. The data cleaning steps include:

(1) Remove noise, such as special characters and punctuations.

(2) Remove numbers.

(3) Remove stopwords by using the [stopwords list](https://gist.github.com/sebleier/554280).

(4) Lowercase all texts

(5) Stemming.

(6) Lemmatization.

```python
# Write your code here

import pandas as pd
df = pd.read_csv("data.csv")
df


```

```python
df['Number of Characters'] = df['reviews'].str.len()
df
```

```python
df['Numbers of numerics'] = df['reviews'].apply(lambda x: len([x for x in x.split() if x.isdigit()]))
df
```

```python
df['digits removal'] = df['reviews'].apply(lambda x: ''.join([i for i in x if not i.isdigit()]))
df
```

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
stop = stopwords.words('english')
df['Stopwords Removal'] = df['reviews'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
df
```

```python
df['Lower Case'] = df['reviews'].apply(lambda x: " ".join(x.lower() for x in x.split()))
df
```

```python
from nltk.stem import PorterStemmer
st = PorterStemmer()
df['Stemming'] = df['Lower Case'].apply(lambda x: " ".join([st.stem(word) for word in x]))
df
```

```python
df['Removal of Punctuation'] = df['Lower Case'].str.replace('[^\w\s]','')
df
```

```python
from textblob import Word
import nltk
nltk.download('wordnet')
df['Lemmatization'] = df['Removal of Punctuation'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
df
```

```python
df.to_csv('cleaned_textdata.csv',index=False)
```

**Write your explanations of the constituency parsing tree and dependency parsing tree here (Question 3-2):**

Natural language processing uses two techniques to examine a sentence's syntactic structure: dependency parsing and constituent parsing. Constituency parsing generates trees that follow a context-free grammar and contain a syntactical representation of a sentence. This kind of tree breaks the sentence up into constituents, which are smaller phrases that fall under a particular grammatical category. For instance, in English, the words "eat a pizza" and "go to the beach" are verb phrases, yet "a dog," "a computer on the table," and "the nice sunset" are all noun phrases. Typically, a constituency parser produces a parse tree as its output, which depicts the sentence's hierarchical structure.

Dependency parsing, on the other hand, is the process of identifying a sentence's grammatical structure by listing each word as a node and showing linkages to its dependents. Every word in the tree is represented by a vertex; words that are reliant on their parents are represented as child nodes; and relationships are used to designate edges. In "John sees Bill," for instance, "sees" is the root of the tree, while "John" and "Bill" are its dependents. Certain downstream activities, such as information extraction and question answering, might benefit more from dependency parsing. For instance, it simplifies the process of extracting subject-verb-object triples, which are frequently a sign of the semantic relationships between predicates.

Constituency parsing and dependency parsing are both relevant to computational linguistics, but their underlying grammars and underlying assumptions lead to distinct trees. Constituency parsing concentrates on the sentence's hierarchical structure, whereas dependency parsing concentrates on its linear structure. Which parser to choose depends on the particular job at hand.
